### locality_name - название населённого пункта

```python
data['locality_name'].isna().mean()*100
```

```python
data.loc[data['locality_name'].isna()].sample(5)
```

т.к. пропусков мало то можно удалить строки с пропусками

```python
data.dropna(subset=['locality_name'], inplace=True)
data['locality_name'].isna().mean()
```

взглянем на названия

```python
data['locality_name'].unique()
```

```python
len(data['locality_name'].unique())
```

проведем замену "поселок" на "посёлок"

```python
# на входе наименование населенного пнкта, 
# если в названии присутсвует "поселок" то заменяем на "посёлок"
def replace_e(name):
    if 'поселок' in name:
        return name.replace('поселок', 'посёлок')
    else:
        return name     
```

```python
data['locality_name'] = data['locality_name'].apply(replace_e)
len(data['locality_name'].unique())
```

### этаж квартиры

категоризируем этажы квартир, результат в новый столбец

```python
# на входе строка, на выходе категория
def to_floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'    
```

```python
data['floor_group']=data.apply(to_floor_group, axis=1)
```

```python
data['floor_group'].value_counts()
```
